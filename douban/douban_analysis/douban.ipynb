```python
import numpy as np
import pandas as pd
from pandas import Series, DataFrame
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
movie = pd.read_csv('movie.csv')
```

```python
movie
```

```python
x = movie['score'].unique()
```

```python
y = movie['name']
```

```python
y = movie['score'].value_counts()
```

```python
y
```

```python
# 250部电影中每个分数的电影数量比较
y.plot.bar(color='b', label = "number")
plt.legend(loc='upper right')
```

```python
y_year = movie['year'].value_counts()
```

```python
y_year = y_year.sort_index()
```

```python
# 总结250部电影中分别分布在哪年
plt.figure(figsize=(12,9))
y_year.plot.bar(color='r', label = "number")
plt.legend(loc='upper left')
```

```python
y_area = movie['area']
```

```python
y_area
```

```python
for i in range(len(y_area)):
    area_list = y_area[i].split(" ")
    if len(area_list) > 1:
        y_area[i] = area_list[0]
```

```python
y_area = y_area.value_counts()
```

```python
y_area
```

```python
# 总结250部电影中主要在哪些国家
plt.figure(figsize=(12,9))
y_area.plot.bar(color='r', label = "number")
plt.legend(loc='upper right')
```
